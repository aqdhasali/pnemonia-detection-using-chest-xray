### Convolution Neural Network

As discussed in Part  A, CNNs have been popular due to their improved performance in image classification. The convolutional layers helps to extract the spacial and temporal features of an image with the help of filters(Rahman, T., et al., 2020).  As the layers have a weight sharing technique it helps in reducing computation efforts.

CNNs excel at automatically identifying features and spatial relationships in 2D images, making them particularly effective for tasks like detecting pneumonia, breast cancer, and more. They remove the need for manual feature extraction, which simplifies image based analysis. However, CNNs come with challenges, such as the need for large amounts of labeled training data something often limited in medical contexts and their high computational demands.

In practice, a CNN model could be trained on a dataset of labeled chest X-ray images, using pixel data as inputs and producing a binary result to indicate whether penumonia is present in an individual. While CNNs are powerful tools for capturing intricate patterns and aiding in early detection, itâ€™s essential to account for their resource requirements and the need for extensive datasets in image classifcation for medical applications.





### Transfer Learning


CNNs tend to perform better with larger datasets compared to smaller ones. In cases where the dataset is limited, transfer learning can be an effective approach for applying CNNs. Various field applications has shown succesful results by using Transfer learning such as medical, manufacturing and baggage screening applications. Transfer learning reduces the need of having larger datasets and also long trainining periods (Rahman, T., et al., 2020).


The main advantage of transfer learning is its ability to use pre-trained models, such as Convolutional Neural Networks (CNNs), that have been trained on large and diverse datasets like ImageNet. By adapting these models for specific tasks, transfer learning reduces the reliance on extensive labeled medical data, which is especially helpful when working with smaller datasets. However, there are some challenges to consider. One issue is that the assumption that the knowledge gained from the source task is fully applicable to the target task may not always be true, which can lead to biased results. Additionally, there's a risk of transferring biases from the source domain into the medical domain, which can be problematic.


### Support Vector Machines

Support Vector Machines (SVM) is considered as a powerful computational model for image classification tasks. It is a supervised learning technique which is applied in the field of classifcation and regression.Support Vector Machines (SVMs) provide a strong method for detecting pneumonia through image classification, particularly effective in dealing with high-dimensional data.

SVMs are well-suited for complex data structures, making them ideal for classifying features extracted from chest X-ray images. Their flexibility is highlighted by their ability to utilize various kernel functions to project data into higher dimensions. However, SVMs also have some challenges. While they can be very effective, they may struggle with large datasets, leading to longer training times. Additionally, SVMs are sensitive to noise in the data and require careful selection of kernel functions to effectively manage non-linear relationships.

In practical applications, SVMs can be used to classify features extracted from chest X-rays, aiding in the detection of pneumonia. In this scenario, the input data would consist of these features, while the output would be a binary classification indicating whether pneumonia is present or not. Despite their strengths in handling high-dimensional data, the sensitivity of SVMs to noise and the potential issues with large datasets need to be taken into account when designing pneumonia detection systems.

```python
# process the image data and prepare it for input
def img_Data(path_dir,target_size,batch,class_list,preprocessing):
  if preprocessing:
    generate_object = ImageDataGenerator(preprocessing_function=preprocessing)
  else:
    generate_object  = ImageDataGenerator()

  return (generate_object.flow_from_directory(path_dir,
                                                                                        target_size = target_size,
                                                                                        batch_size = batch,
                                                                                        class_mode = 'sparse',
                                                                                        classes = class_list,
                                                                                        shuffle = True))
```

```python
#make the image dimensions same
DATA_TRAINING_PATH = 'train'
BATCH_SIZE = 16
IMG_HEIGHT, IMG_WIDTH = 224, 224
IMG_SIZE  = (IMG_HEIGHT, IMG_WIDTH, 3)

DATA_TESTING_PATH  = 'test'
BATCH_SIZE = 16
IMG_HEIGHT, IMG_WIDTH = 224, 224
IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH, 3)
```

## Data Training

```python
# split the data for training and testing (80-20)
train_data = tf.keras.utils.image_dataset_from_directory(
    DATA_TRAINING_PATH,
    label_mode='categorical',
    color_mode = 'rgb',
    validation_split = 0.2,
    subset="training",
    seed=42,
    shuffle=True,
    image_size=(IMG_HEIGHT,IMG_WIDTH),
    batch_size=BATCH_SIZE
)

test_data = tf.keras.utils.image_dataset_from_directory(
    DATA_TESTING_PATH,
    label_mode='categorical',
    color_mode='rgb',
    validation_split=0.2,
    subset="validation",
    seed=42,
    shuffle=False,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE
)

# Extract test data labels for evaluation
test_labels = []
for images, labels in test_data:
    test_labels.extend(labels.numpy())

test_labels = np.array(test_labels)
```

```python
# extract class names
class_names = train_data.class_names

# optimize
train_data = train_data.prefetch(tf.data.AUTOTUNE)
test_data = test_data.prefetch(tf.data.AUTOTUNE)

print(class_names)
```

```python
from tensorflow import keras
```

# Initialize the pretrained ResNET model

```python
from tensorflow.keras.applications import ResNet50V2
```

```python
#Freeze the base model - basically we are not training the base model because it requires a lot of time
base_model = ResNet50V2(
    include_top=False, # Exclude ImageNet classifier at the top
    weights='imagenet',
    input_shape=IMG_SIZE
    )

base_model.trainable = False
```

# Data Augmentation

```python
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomTranslation, RandomContrast
data_augmentation = keras.Sequential(
    [RandomFlip('horizontal'),
     RandomRotation(factor=(-0.025, 0.025)),
     RandomTranslation(height_factor=0.1, width_factor=0.1),
     RandomContrast(factor=0.1)
     ])



# Create new model on top
inputs = keras.Input(shape=IMG_SIZE)
x = data_augmentation(inputs)

x = keras.applications.resnet_v2.preprocess_input(x)
x = base_model(x, training=False)
```

```python
# adding the classification head on top of the ResNet Model
from tensorflow.keras.layers import GlobalAveragePooling2D, BatchNormalization, Dropout, Flatten, Dense
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)  # Regularize with dropout

outputs = Dense(len(class_names), activation="softmax", name="pred")(x)

resnet_model = keras.Model(inputs, outputs)



resnet_model.summary()
```

```python
# training the resnet model
from tensorflow.keras.metrics import *
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

metrics = [CategoricalAccuracy(), Precision(), Recall()]

resnet_model.compile(
    loss = CategoricalCrossentropy(),
    optimizer = Adam(learning_rate=0.001),
    metrics = metrics
    )

elst = callbacks.EarlyStopping(monitor='val_loss',patience=25,mode='min')
save_ck = callbacks.ModelCheckpoint('resnet-model-23-10-18.keras',save_best_only=True,monitor='val_loss',mode= 'min')


resnet_history = resnet_model.fit(
      train_data,
      epochs=10,
      validation_data=test_data,
      callbacks=[elst],
      verbose=1
      )



resnet_model.save('AICWModel.keras')



result = resnet_model.evaluate(test_data)
result
```

 The ResNet model was trained for 10 epochs and it has shown successful results. The final categorical accuracy of the model is around 99.1% on the validation set, which indicates that the model has correctly classified the images into their respective classes with high accuracy. The precision metrics and recall metrics has shown 98.1% which further elaborates the ability of the model to identify positive instances and minimizing false results. The training and validation curves indicate that the model hasn't overfitted, as both sets showed a consistent improvement in accuracy. However, it's important to evaluate the specific needs and context of the application to decide if this performance is adequate. In the evaluation phase of this research we  explore further on the image classifcation model that was trained for pneumonia detection using chest x-rays.
```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from google.colab import files  # Only for Google Colab users

# Function to create image data generator
def img_Data(path_dir, target_size, batch, class_list, preprocessing):
    if preprocessing:
        generate_object = ImageDataGenerator(preprocessing_function=preprocessing)
    else:
        generate_object = ImageDataGenerator()

    return generate_object.flow_from_directory(
        path_dir,
        target_size=target_size,
        batch_size=batch,
        class_mode='sparse',  # Change to 'binary' if you're doing binary classification
        classes=class_list,
        shuffle=True
    )

# Generate training and test data
generate_train_data = img_Data("train", (224, 224), 32, os.listdir("train"), preprocess_input)
generate_test_data = img_Data("test", (224, 224), 32, os.listdir("test"), preprocess_input)

# Load the MobileNetV2 model
base_model = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling=None
)

# Freeze the base model
base_model.trainable = False

# Build the model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')  # Change to Dense(num_classes, activation='softmax') for multi-class
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # Change if needed
              metrics=['accuracy'])

# Train the model for 3 epochs
history = model.fit(
    generate_train_data,
    epochs=10,
    validation_data=generate_test_data  # Optional for validation during training
)

# Save the model in HDF5 format
model.save('mobilenet_model.h5')

# Download the saved model (for Google Colab users)
files.download('mobilenet_model.h5')

# Evaluate the model on test data
result_mobilenetv2 = model.evaluate(generate_test_data)

# Print evaluation results
print("MobileNetV2 Evaluation Results:")
print("Loss:", result_mobilenetv2[0])
print("Accuracy:", result_mobilenetv2[1])
```

```python
#mobilenetV2 model
loss_train_curve = history.history["loss"]
loss_val_curve = history.history["val_loss"]
plt.plot(loss_train_curve, label = "Train")
plt.plot(loss_val_curve, label = "Validation")
plt.legend(loc = 'upper right')
plt.title("MobileNetV2's Loss")
plt.show()

#mobilenetV2 model
accuracy_train_curve = history.history["accuracy"]
accuracy_val_curve = history.history["val_accuracy"]

plt.plot(accuracy_train_curve, label="Train")
plt.plot(accuracy_val_curve, label="Validation")
plt.legend(loc='lower right')
plt.title("MobileNetV2's Accuracy")
plt.show()
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.efficientnet import preprocess_input
from google.colab import files  # Only for Google Colab users

# Function to create image data generator
def img_Data(path_dir, target_size, batch, class_list, preprocessing):
    if preprocessing:
        generate_object = ImageDataGenerator(preprocessing_function=preprocessing)
    else:
        generate_object = ImageDataGenerator()

    return generate_object.flow_from_directory(
        path_dir,
        target_size=target_size,
        batch_size=batch,
        class_mode='sparse',
        classes=class_list,
        shuffle=True
    )

# Generate training and test data
generate_train_data = img_Data("train", (224, 224), 32, os.listdir("train"), preprocess_input)
generate_test_data = img_Data("test", (224, 224), 32, os.listdir("test"), preprocess_input)

# Load the EfficientNetB0 model
base_model = EfficientNetB0(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling=None
)

# Freeze the base model
base_model.trainable = False

# Build the model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # Change if needed
              metrics=['accuracy'])

# Train the model for 10 epochs
history = model.fit(
    generate_train_data,
    epochs=10,
    validation_data=generate_test_data  # Optional for validation during training
)

# Save the model in HDF5 format
model.save('efficientnet_model.keras')

# Download the saved model (for Google Colab users)
files.download('efficientnet_model.keras')

# Evaluate the model on test data
result_efficientnet = model.evaluate(generate_test_data)

# Print evaluation results
print("EfficientNetB0 Evaluation Results:")
print("Loss:", result_efficientnet[0])
print("Accuracy:", result_efficientnet[1])
```
